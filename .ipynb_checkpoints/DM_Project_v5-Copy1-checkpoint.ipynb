# Definici√≥n de Funciones

```python
def limpia_fechas(fecha):
    fecha_lista = [int(element) for element in fecha.split('-')]
    return datetime.datetime(fecha_lista[0], fecha_lista[1], fecha_lista[2])
```

```python
def limpia_horas(hora):
    hora_lista = [int(element) for element in hora.split(':')]
    return datetime.time(hora_lista[0], hora_lista[1])
```

```python
def extract_text(mess):
    return ''.join(c for c in mess if c not in emoji.UNICODE_EMOJI)
```

```python
df_texto['fecha'] = df_texto['fecha'].map(limpia_fechas)
df_texto['hora'] = df_texto['hora'].map(limpia_horas)
```

# Cluster

```python
X = pd.DataFrame({'minutos':x, 'num_mensajes':y})

X.shape
```

X.to_csv('../../Data/WhatsApp/Puntos_acomodados.csv', index=False)

```python
X_result = X.copy()
```

```python
X_result.head()
```

```python
lst_in = []
for k in range(2,11):
    km = KMeans(n_clusters=k)
    km.fit(X)
    lst_in.append(km.inertia_)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(range(2,11),lst_in, marker='o')
plt.show()
```

Tomando la regla del *codito*, el n√∫mero de cl√∫sters deber√≠a ser 4, pero el problema necesita tener 3 o 5 grupos, as√≠ que tomaremos 5 como el n√∫mero de cl√∫sters.


```python
n=5
agg = AgglomerativeClustering(n_clusters=n)
km = KMeans(n_clusters=n, random_state=123)
gmm = GaussianMixture(n_components=n, random_state=123)

km.fit(X)
gmm.fit(X)
```

```python
X_result['cl_agg'] = agg.fit_predict(X)
X_result['cl_km'] = km.predict(X)
X_result['cl_gmm'] = gmm.predict(X)
```

```python
sns.lmplot(x = 'minutos', y = 'num_mensajes', data=X_result, height=8, aspect=2.2, fit_reg=False, hue='cl_km')
plt.show()
```

```python
sns.lmplot(x = 'minutos', y = 'num_mensajes', data=X_result, height=8, aspect=2.2, fit_reg=False, hue='cl_agg')
plt.show()
```

```python
sns.lmplot(x = 'minutos', y = 'num_mensajes', data=X_result, height=8, aspect=2.2, fit_reg=False, hue='cl_gmm')
plt.show()
```

```python
xs = {'x' + str(i):[] for i in range(1, 12)}
list_cluster = []
list_fechai = []
list_fechaf = []
list_num_whats = []
for i in range(5):
    list_filter = X_result[X_result['cl_km'] == i]['minutos'].to_list()
    globals()['kmcl_' + str(i)] =  df_texto[df_texto['lab_minutes'].apply(lambda x: x in list_filter)]
    list_cluster.append('Cluster ' + str(i))
    min_min = globals()['kmcl_' + str(i)]['lab_minutes'].min()
    max_min = globals()['kmcl_' + str(i)]['lab_minutes'].max()
    list_fechai.append(globals()['kmcl_' + str(i)][globals()['kmcl_' + str(i)]['lab_minutes'] == min_min]['hora'].unique()[0].strftime("%H:%M"))
    list_fechaf.append(globals()['kmcl_' + str(i)][globals()['kmcl_' + str(i)]['lab_minutes'] == max_min]['hora'].unique()[0].strftime("%H:%M"))
    list_num_whats.append(len(globals()['kmcl_' + str(i)]))
    for k in xs.keys():
        try:
            xs[k].append(globals()['kmcl_' + str(i)].groupby(['persona']).count()['mensaje'].loc[k])
        except:
            xs[k].append(0)

dict_columns = dict({'N√∫m Cluster': list_cluster, 'Desde': list_fechai, 'Hasta': list_fechaf}, **dict(xs,  **{'Total Mensajes': list_num_whats}))
clusters_resum = pd.DataFrame(dict_columns)
```
